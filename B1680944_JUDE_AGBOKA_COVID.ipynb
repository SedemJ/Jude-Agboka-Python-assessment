```python
#To check if the given date is between the given date range in the database
import datetime

def check_date():

    date_input = input ('Date: ')
    date_tokens = date_input.split('/')
    day = int(date_tokens[0])
    month = int(date_tokens[1])
    year = int(date_tokens[2])
    date_input = datetime.date(year, month, day)
    if date_input > datetime.date(2020, 1, 1) and date_input < datetime.date(2020, 12, 31):
        print('valid')
        
    else: 
        print("The date that you enter is not valid, please input a date in between 1/1/2020 and 31/12/2020 ")
        
check_date()
```
